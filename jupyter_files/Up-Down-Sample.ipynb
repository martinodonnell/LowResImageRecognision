# Down sample

```python
#Down samples images from stanford to (122,122) to match that of boxcars
import os
import cv2
import numpy as np
from matplotlib import pyplot as plt
from tqdm import tqdm
```

```python


folder="/Users/martinodonnell/Documents/uni/fourth_year/4006/LowResImageRecognision/data/StanfordCars/"
for subdir, dirs, files in os.walk(folder):
    for file in tqdm(files):
        #print os.path.join(subdir, file)
        filepath = subdir + os.sep + file

        if filepath.endswith(".jpg"):
            img = cv2.imread(filepath)
            res = cv2.resize(img, dsize=(122,122), interpolation=cv2.INTER_CUBIC)
            cv2.imwrite(filepath[:-4] + "_ds.jpg", res)
                
```

# Get average resolution

```python
from PIL import Image

# folder='/Users/martinodonnell/Documents/uni/fourth_year/4006/LowResImageRecognision/data/BoxCars/images'
folder="/Users/martinodonnell/Documents/uni/fourth_year/4006/LowResImageRecognision/data/StanfordCars/"
width = 0
height=0
count=0
for subdir, dirs, files in os.walk(folder):
    for file in tqdm(files):
        #print os.path.join(subdir, file)
        filepath = subdir + os.sep + file

        if filepath.endswith(".jpg") and not filepath.endswith("ds.jpg"):
            # open image for reading in binary mode
            im = Image.open(filepath)
            w, h = im.size
            width += w
            height+= h
            count+=1
#             print("The resolution of the image is",w,"x",h,'-',file)
print(w,h)
                
```
